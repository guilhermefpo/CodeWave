#Versão com arquivo .CSV (Comma Separated Values):

```python
#instalação e declaração de bibliotecas
!pip install pandas openpyxl

import pandas as pd
from google.colab import files

 #Upload do arquivo (.CSV)
uploaded = files.upload()
file_path = list(uploaded.keys())[0]

#Leitura do arquivo
df = pd.read_csv(file_path, sep= ";", encoding = "latin1")

#Remover linhas vazias
df = df.dropna(how="all")

# 4. Detectar automaticamente a(s) coluna(s) que contém o código de município 3549904
mask = df.apply(lambda row: row.astype(str).str.contains("3549904", na=False)).any(axis=1)

# 5. Filtrar apenas essas linhas
df_filtrado = df[mask]

# 6. Salvar em CSV (mudar "municipio_3549904" para não ter vários arquivos com o mesmo nome com dados diferentes)
output_file = "municipio_3549904.csv"
df_filtrado.to_csv(output_file, sep= ";", index=False, encoding="utf-8-sig")

# 7. Download do arquivo tratado
files.download(output_file)
```

#Versão com arquivo .XLSX (Excel)

a principal mudança é no pd.read que passa a ter a extenção "excel" e as definições "file_path" e "sheet_name" para ler as planilhas do arquivo xls ou xlsx

```python
#instalação e declaração de bibliotecas
!pip install pandas openpyxl

import pandas as pd
from google.colab import files

 #Upload do arquivo (.xlsx)
uploaded = files.upload()
file_path = list(uploaded.keys())[0]

#Leitura do arquivo
df = pd.read_excel(file_path, sheet_name="")

#Remover linhas vazias
df = df.dropna(how="all")

# 4. Detectar automaticamente a(s) coluna(s) que contém o código de município 3549904
mask = df.apply(lambda row: row.astype(str).str.contains("3549904", na=False)).any(axis=1)

# 5. Filtrar apenas essas linhas
df_filtrado = df[mask]

# 6. Salvar em CSV (mudar "municipio_3549904" para não ter vários arquivos com o mesmo nome com dados diferentes)
output_file = "municipio_3549904.csv"
df_filtrado.to_csv(output_file, sep= ";", index=False, encoding="utf-8-sig")

# 7. Download do arquivo tratado
files.download(output_file)
```
