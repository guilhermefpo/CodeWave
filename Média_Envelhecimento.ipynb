<h3> Média de Idosos a cada 100 Habitantes (2010 / 2022) </h3>

```python
import pandas as pd
import plotly.graph_objects as go
```

```python
# Dados
dados = {
    "Ano": [2010, 2022],
    "São Paulo (Estado)": [54, 96],
    "São José dos Campos (Município)": [45, 91]
}
df = pd.DataFrame(dados)

# Graficos
fig = go.Figure()

# Grafico São Paulo
fig.add_trace(go.Bar(
    x=df["Ano"],
    y=df["São Paulo (Estado)"],
    name="São Paulo (Estado)",
    text=df["São Paulo (Estado)"],
    textposition="outside",
    marker_color="Red",
    visible=True
))

# Grafico SJC
fig.add_trace(go.Bar(
    x=df["Ano"],
    y=df["São José dos Campos (Município)"],
    name="São José dos Campos (Município)",
    text=df["São José dos Campos (Município)"],
    textposition="outside",
    marker_color="#21ABCD",
    visible=True
))

fig.update_layout(
    barmode="group",
    title="Índice de Envelhecimento (2010 vs 2022)",
    xaxis_title="Ano",
    yaxis_title="Índice de Envelhecimento (cada 100 Habitantes)",
    template="plotly_white",
    legend_title="Localidade"
)

fig.show()
```
